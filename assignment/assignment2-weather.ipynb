### Import libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

### Import dataset

```python
file_path = 'weatherreadings1.csv'
data = pd.read_csv(file_path)
```

#### Show first rows of the data

```python
print(data.head())
```

#### Show column 'dryBulbTemperature_Celsius'

```python
print(data['dryBulbTemperature_Celsius'])
```

#### Check for null values in the 'dryBulbTemperature_Celsius' column

```python
print(data['dryBulbTemperature_Celsius'].isnull().sum())
```

#### Convert the date column to datetime format if applicable

```python
data['reportEndDateTime'] = pd.to_datetime(data['reportEndDateTime'])
```

#### Set the date as the index for plotting

```python
data.set_index('reportEndDateTime', inplace=True)
```

#### Plotting the temperature over time

```python
plt.figure(figsize=(10, 4))
plt.plot(data.index, data['dryBulbTemperature_Celsius'], color='green', linewidth=2)
plt.title('Dry Bulb Temperature Over Time', fontsize=16)
plt.xlabel('Date', fontsize=10)
plt.ylabel('Temperature (C)', fontsize=10)
plt.grid(True)
plt.xticks(rotation=45)
plt.tight_layout()

plt.show()
```
