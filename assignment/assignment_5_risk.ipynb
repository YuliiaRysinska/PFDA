#### Import necessary libraries

```python
import numpy as np
import matplotlib.pyplot as plt
```

#### Create parameters

```python
num_rounds = 1000
```

#### Define function to simulate battle round

```python
def battle_round():
    # Roll dice for attacker (3 dice) and defender (2 dice)
    attacker_rolls = np.sort(np.random.randint(1, 7, 3))[::-1]  # Sort in descending order
    defender_rolls = np.sort(np.random.randint(1, 7, 2))[::-1]  # Sort in descending order

    # Initialize troop losses
    attacker_losses = 0
    defender_losses = 0

    # Compare the highest roll (1st pair)
    if attacker_rolls[0] > defender_rolls[0]:
        defender_losses += 1
    else:
        attacker_losses += 1

    # Compare second highest roll (2nd pair)
    if attacker_rolls[1] > defender_rolls[1]:
        defender_losses += 1
    else:
        attacker_losses += 1

    return attacker_losses, defender_losses
    
```

#### Run 

```python
results = [battle_round() for _ in range(num_rounds)]
attacker_losses = [result[0] for result in results]
defender_losses = [result[1] for result in results]
```

#### Count the outcomes

```python
outcomes, counts = np.unique([(a, d) for a, d in results], return_counts=True, axis=0)
```

#### Plotting results

```python
fig, ax = plt.subplots(figsize=(8, 4))
ax.bar([f"A:{a} - D:{d}" for a, d in outcomes], counts, color="g")
ax.set_xlabel("Outcome")
ax.set_ylabel("Frequency")
ax.set_title("Distribution of Outcomes in Risk Game")

plt.xticks(rotation=45)
plt.tight_layout()
plt.show()
```
