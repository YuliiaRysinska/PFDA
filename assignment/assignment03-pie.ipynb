#### Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
```

#### Load dataset and show first rows

```python
url = 'https://drive.google.com/uc?id=1AWPf-pJodJKeHsARQK_RHiNsE8fjPCVK&export=download'
data = pd.read_csv(url)

data.head()
```

 #### Extract domain email

```python
data['domain'] = data['Email'].apply(lambda x: x.split('@')[1])

domain_count = data['domain'].value_counts()
```

#### Create Pie Chart

```python
plt.figure(figsize=(8, 6))
plt.pie(domain_count, labels=domain_count.index, autopct='%1.1f%%', startangle=140)
plt.title('Email Domains')
plt.axis('equal')
plt.show()
```
