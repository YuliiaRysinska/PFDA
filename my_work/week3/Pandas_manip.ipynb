```python
people_df = pd.read_csv(people_filename)
people_df.head(3)
```

to get the email domains

```python
people_df['domain'] = people_df['Email'].str.split('@').str[1]
people_df.head(3)

domain_count = people_df['domain'].value_counts()
```

How would you make a pie chart for the email domains

```python
plt.figure(figsize=(8, 6))
plt.pie(domain_count, labels=domain_count.index, autopct='%1.1f%%', startangle=140)
plt.title('Email Domains')
plt.axis('equal')
plt.show()
```
